```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
K = float(input("Strike Price : "))
P = float(input("Premium : "))
minS = K - 50
maxS = K + 50
spotprices = np.linspace(minS, maxS, 50)
```

```python
buy_call_option = np.maximum(spotprices - K, 0) - P
sell_call_option = P - np.maximum(spotprices - K, 0)
buy_put_option = np.maximum(K - spotprices, 0) - P
sell_put_option = P - np.maximum(K - spotprices, 0)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(spotprices, buy_call_option, label="Buy Call", color='green')
plt.plot(spotprices, sell_call_option, label="Sell Call", color='red')
plt.plot(spotprices, buy_put_option, label="Buy Put", color='blue')
plt.plot(spotprices, sell_put_option, label="Sell Put", color='pink')

plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(K, color='gray', linestyle='--', label='Strike Price')

plt.title("Option Payoff")
plt.xlabel("Spot Price")
plt.ylabel("Profit&Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#OBSERVATION

#BUY CALL: As the spot price increases till 806.5 which is breakeven point buyer incures loss after that profit in increasing and if the
#          spot price go below 800 the loss incurred is only the premium paid 

#SELL CALL:  As the spot price decreases seller incures profit equals to the premium paid after that loss is increasing 
# similar concept for BUY PUT AND SELL PUT
```
